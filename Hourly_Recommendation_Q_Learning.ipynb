```python
import pandas as pd
import numpy as np
import random
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder

# --- Load data ---
# df = pd.read_csv("your_data.csv", parse_dates=['date'])

# --- Encode carbon_level ---
le = LabelEncoder()
df['carbon_level_enc'] = le.fit_transform(df['carbon_level'])  # low=1, medium=2, high=3 assumed

# --- Define Action Items ---
actions = [
    "No action",  # 0

    # Energy Efficiency
    "Turn off unused appliances",  # 1
    "Delay high-energy tasks to off-peak hours",  # 2
    "Use smart lighting schedule",  # 3
    "Reduce heating/cooling during mild weather",  # 4
    "Switch off standby devices (TV, routers, etc.)",  # 5
    "Use smart thermostats to optimize temperature",  # 6
    "Upgrade to a smart meter to track energy consumption",  # 15
    "Install motion sensors for lights in unused rooms",  # 16

    # Carbon + Sustainability
    "Delay heavy usage during high carbon periods",  # 7
    "Recommend green energy subscription (solar, wind)",  # 8
    "Use natural ventilation instead of air conditioning",  # 9
    "Install solar panels",  # 17

    # Lifestyle-tailored Tips
    "Encourage weekend laundry/dishwashing (lower tariffs)",  # 10
    "Use microwave instead of oven for small meals",  # 11
    "Promote group activities outdoors to save electricity indoors",  # 12

    # Appliance Upgrade
    "Try upgrading to energy-efficient appliances (A++ rating)",  # 13
    "Do maintenance check for heating/cooling systems",  # 14
    "Upgrade windows to double-glazing for better insulation",  # 18
    "Use low-energy ceiling fans for cooling instead of air conditioning",  # 19
]

# --- Initialize Q-Tables ---
carbon_q_table = defaultdict(lambda: np.zeros(len(actions)))
energy_q_table = defaultdict(lambda: np.zeros(len(actions)))
active_q_table = defaultdict(lambda: np.zeros(len(actions)))

# --- Helper: discretize state ---
def get_state(row):
    return (
        int(row['hour']),
        int(row['day_of_week']),
        int(row['predicted_is_active']),
        int(row['carbon_level_enc']),
        int(row['Appliances'] > row['power_threshold']),
        int(row['avg_temp'] >= 15),  # hot days
        int(row['avg_temp'] < 15),  # cold days
    )

def carbon_reward(row, action):
    reward = 0
    if action == 8 and row['carbon_level_enc'] >= 2:  # Recommend green energy subscription (solar, wind)
        reward += 2  # Carbon savings from renewable energy sources
    if action == 7 and row['carbon_level_enc'] >= 2:  # Delay heavy usage during high carbon periods
        reward += 2  # Carbon reduction by avoiding peak carbon times
    if action == 17 and row['carbon_level_enc'] >= 2:  # Install solar panels
        reward += 3  # High impact on reducing carbon footprint from renewable energy
    return reward

def active_reward(row, action):
    reward = 0
    if action == 1 and row['predicted_is_active'] == 0:  # Turn off unused appliances if user is away
        reward += 2  # Energy saved when nobody is around
    if action == 6 and row['predicted_is_active'] == 1:  # Smart thermostats optimize temperature when active
        reward += 1  # Helps when someone is home
    if action == 5 and row['predicted_is_active'] == 0:  # Switch off standby devices (TV, routers, etc.) when no one is home
        reward += 2  # Energy saved by turning off devices in standby mode
    if action == 9 and row['predicted_is_active'] == 1:  # Use natural ventilation instead of air conditioning when someone is home
        reward += 1  # Helps save energy when the house is occupied
    return reward

def energy_reward(row, action):
    reward = 0
    if action == 2 and (row['hour'] in [18, 19, 20]):  # Delay high-energy tasks to off-peak hours (evening peak)
        reward += 1  # Off-peak scheduling reduces energy costs and grid strain
    if action == 1 and row['Appliances'] > 300:  # Turn off unused appliances if high energy usage
        reward += 2  # Reduce consumption by turning off energy-hungry devices
    if action == 4 and (row['T_out'] < 10 or row['T_out'] > 20):  # Reduce heating/cooling during mild weather
        reward += 1  # Saves energy by optimizing HVAC for temperature conditions
    if action == 6:  # Use smart thermostats to optimize temperature
        reward += 2  # Smart thermostats are efficient for both active and inactive times
    if action == 8 and row['T_out'] < 10:  # Recommend green energy subscription (solar, wind) when it's cold
        reward += 1  # Greater impact on energy savings in colder weather
    if action == 15:  # Upgrade to a smart meter to track energy consumption
        reward += 1  # Helps users track energy usage more efficiently
    if action == 16:  # Install motion sensors for lights in unused rooms
        reward += 1  # Automatically turning off lights in unoccupied rooms reduces energy waste
    return reward


# --- Q-learning hyperparams ---
epsilon = 0.2
alpha = 0.1
gamma = 0.9

# --- Training loop (multi-agent learning) ---
for epoch in range(100):
    for idx, row in df.iterrows():
        state = get_state(row)

        # --- Carbon Agent ---
        if random.random() < epsilon:
            action_idx = random.randint(0, len(actions) - 1)
        else:
            action_idx = np.argmax(carbon_q_table[state])
        reward = carbon_reward(row, action_idx)
        next_max = np.max(carbon_q_table[state])
        carbon_q_table[state][action_idx] += alpha * (reward + gamma * next_max - carbon_q_table[state][action_idx])

        # --- Energy Agent ---
        if random.random() < epsilon:
            action_idx = random.randint(0, len(actions) - 1)
        else:
            action_idx = np.argmax(energy_q_table[state])
        reward = energy_reward(row, action_idx)
        next_max = np.max(energy_q_table[state])
        energy_q_table[state][action_idx] += alpha * (reward + gamma * next_max - energy_q_table[state][action_idx])

        # --- Active Agent ---
        if random.random() < epsilon:
            action_idx = random.randint(0, len(actions) - 1)
        else:
            action_idx = np.argmax(active_q_table[state])
        reward = active_reward(row, action_idx)
        next_max = np.max(active_q_table[state])
        active_q_table[state][action_idx] += alpha * (reward + gamma * next_max - active_q_table[state][action_idx])
```

```python
# --- Multi-Agent Fusion (List of Suggestions) ---
def fused_recommend(row, w_carbon=0.3, w_energy=0.4, w_active=0.3):
    state = get_state(row)
    carbon_values = carbon_q_table[state]
    energy_values = energy_q_table[state]
    active_values = active_q_table[state]
    fused_q = w_carbon * carbon_values + w_energy * energy_values + w_active * active_values

    top_indices = np.argsort(fused_q)[-5:]
    return [actions[i] for i in reversed(top_indices)]

# --- Example Usage ---
sample_df1 = df.sample(3)
for _, row in sample_df.iterrows():
    recs = fused_recommend(row)
    print(f"📅 Date: {row['date']}")
    print("📋 Recommendations:")
    for r in recs:
        print(f" - {r}")
    print("\n" + "-"*50 + "\n")
```

```python
# --- Multi-Agent Fusion (Ranked List) ---

def fused_recommend2(row, w_carbon=0.3, w_energy=0.4, w_active=0.3):
    state = get_state(row)
    carbon_values = carbon_q_table[state]
    energy_values = energy_q_table[state]
    active_values = active_q_table[state]

    # Combine Q-values using weights
    fused_q = w_carbon * carbon_values + w_energy * energy_values + w_active * active_values

    # Rank actions by fused Q-values in descending order
    ranked_indices = np.argsort(fused_q)[::-1]  # Sort indices by fused Q-values, highest first

    # Return top 3 ranked actions based on fused Q-values
    return [actions[i] for i in ranked_indices[:5]]

# Example: Get recommendations with ranking
sample_df2 = df.sample(3)
for _, row in sample_df.iterrows():
    recs = fused_recommend(row)
    print(f"📅 Date: {row['date']}")
    print("📋 Ranked Recommendations:")
    for r in recs:
        print(f" - {r}")
    print("\n" + "-"*50 + "\n")
```

```python
rooms = {
    "kitchen": ("T1", "R1"),
    "living_room": ("T2", "R2"),
    "laundry_room": ("T3", "R3"),
    "office_room": ("T4", "R4"),
    "bathroom": ("T5", "R5"),
    "ironing_room": ("T7", "R7"),
    "teenager_room": ("T8", "R8"),
    "parents_room": ("T9", "R9"),
}

room_specific_actions = {
    "kitchen": [
        "Turn off unused appliances",
        "Suggest using microwave instead of oven for small meals",
        "Delay high-energy tasks to off-peak hours",
        "Switch off standby devices (oven, chimney)",
        "Suggest upgrading to energy-efficient appliances (A++ rating)",
        "Use induction cooktops instead of traditional gas burners",
        "Optimize refrigerator usage (set temperature between 37°F and 40°F)",
        "Use energy-efficient lighting (LED bulbs, motion-sensor lights)",
    ],
    "living_room": [
        "Suggest smart lighting schedule",
        "Promote group activities outdoors to save electricity indoors",
        "Switch off standby devices (TV, routers, speaker)",
        "Use smart thermostats to optimize temperature",
        "Unplug electronics when not in use to prevent standby energy consumption",
        "Use programmable thermostats to reduce heating and cooling",
        "Use a dehumidifier to reduce HVAC strain in humid areas",
    ],
    "laundry_room": [
        "Encourage weekend laundry/dishwashing (lower tariffs)",
        "Delay heavy usage during high carbon periods",
        "Suggest upgrading to energy-efficient appliances (A++ rating)",
        "Wash clothes in cold water to reduce energy use from heating water",
        "Avoid overloading the washing machine for optimal performance",
        "Use the dryer’s moisture sensor to stop drying once clothes are dry",
    ],
    "office_room": [
        "Use smart thermostats to optimize temperature",
        "Switch off standby devices (TV, routers)",
        "Suggest smart lighting schedule",
        "Enable power-saving mode on all devices to reduce energy consumption",
        "Adjust screen brightness to optimal levels to save power",
    ],
    "bathroom": [
        "Use natural ventilation instead of fan/air conditioning",
        "Reduce heating/cooling during mild weather",
        "Switch to motion sensors for automatic light control",
        "Install a timer for bathroom fans to avoid unnecessary operation",
        "Use a water heater timer to heat water only when necessary",
    ],
    "ironing_room": [
        "Delay high-energy tasks to off-peak hours",
        "Suggest upgrading to energy-efficient appliances (A++ rating)",
        "Iron in bulk to maximize energy use",
        "Use an iron with a steam burst feature for more efficient ironing",
        "Ensure proper ventilation to avoid overheating the room",
    ],
    "teenager_room": [
        "Turn off unused appliances",
        "Switch off standby devices (TV, speaker)",
        "Encourage using natural light and lighter-colored curtains",
        "Unplug unused electronics like gaming consoles and TVs",
        "Install smart lighting with motion sensors",
        "Use energy-efficient gaming setups",
        "Set a timer for devices like heaters or fans",
    ],
    "parents_room": [
        "Use smart thermostats to optimize temperature",
        "Switch off standby devices (TV, routers)",
        "Delay heavy usage during high carbon periods",
        "Use blackout curtains to prevent heat loss in winter and heat gain in summer",
        "Install a smart thermostat that adjusts temperature based on family routines",
        "Encourage using sleep mode for devices to reduce standby consumption",
        "Opt for low-energy lighting (LEDs or dimmer switches)",
        "Use a smart home system to monitor and control power usage remotely",
    ],
}

def recommend_actions5(row, carbon_q_table, energy_q_table, active_q_table, actions, top_k=3, epsilon=0.1):
    recommendations = []
    for room in rooms:
        state = get_state(row, room)

        # --- Use room-specific actions ---
        available_actions = room_specific_actions.get(room, actions)  # fallback to global if not found

        if state not in carbon_q_table:
            print(f"State {state} not found. Random actions for {room}.")
            # Ensure top_k does not exceed the number of available actions
            top_k_actual = min(top_k, len(available_actions))
            action_indices = np.random.choice(len(available_actions), size=top_k_actual, replace=False)
            selected_actions = [available_actions[i] for i in action_indices]
            recommendations.extend(selected_actions)
        else:
            # Fuse Q-values
            fused_q = fused_q_table(
                carbon_q_table[state],
                energy_q_table[state],
                active_q_table[state]
            )

            # --- Trick: For exploitation, assign random scores if action is unknown ---
            fused_scores = []
            for a in available_actions:
                if a in actions:
                    fused_scores.append(fused_q[actions.index(a)])
                else:
                    fused_scores.append(np.random.rand())  # random small value if not known

            fused_scores = np.array(fused_scores)

            if np.random.rand() < epsilon:
                print(f"Exploration: random actions for {room}.")
                # Ensure top_k does not exceed the number of available actions
                top_k_actual = min(top_k, len(available_actions))
                action_indices = np.random.choice(len(available_actions), size=top_k_actual, replace=False)
            else:
                action_indices = np.argsort(fused_scores)[-top_k:][::-1]

            selected_actions = [available_actions[i] for i in action_indices]
            recommendations.extend(selected_actions)

    return recommendations


# example
sample_df = df.sample(3)

for idx, row in sample_df.iterrows():
    print(f"\n🗓️ Date: {row['date']}")  # if your date column is named 'date'
    # print(f"🔍 Recommendations for Row {idx}:")

    recommendations = recommend_actions5(
        row,
        carbon_q_table,
        energy_q_table,
        active_q_table,
        actions,
        top_k=3
    )

    # Grouping by rooms
    for room_idx, room in enumerate(rooms):
        print(f"\n🏠 Room: {room.replace('_', ' ').title()}")
        for rec in recommendations[room_idx * 3 : (room_idx + 1) * 3]:  # 3 actions per room
            print(f" - {rec}")
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import pickle
from collections import defaultdict

# --- Load Data ---
@st.cache_data
def load_data():
    df = pd.read_csv("your_data.csv", parse_dates=['date'])
    return df

df = load_data()

# --- Load Q-table ---
@st.cache_data
def load_q_table():
    with open("q_table.pkl", "rb") as f:
        q_table_loaded = defaultdict(lambda: np.zeros(len(actions)), pickle.load(f))
    return q_table_loaded

# --- Define actions ---
actions = [
    "No action",
    "Turn off unused appliances",
    "Delay high-energy tasks",
    "Suggest smart lighting schedule",
    "Reduce heating/cooling"
]

q_table = load_q_table()

# --- Helper functions ---
def get_state(row):
    return (
        int(row['hour']),
        int(row['day_of_week']),
        int(row['is_active']),
        int(row['predicted_is_active']),
        int(row['carbon_level_enc']),
        int(row['Appliances'] > row['power_threshold'])
    )

def recommend(row):
    state = get_state(row)
    best_action_idx = np.argmax(q_table[state])
    return actions[best_action_idx]

# --- Sidebar ---
st.sidebar.title("Energy Recommendation System")
num_samples = st.sidebar.slider("Number of Random Samples", 1, 20, 5)

# --- Main View ---
st.title("🏡 Smart Home Energy Recommendations")
st.write("Using RL-based agents to suggest eco-friendly actions.")

sample_df = df.sample(num_samples).copy()

for i, row in sample_df.iterrows():
    rec = recommend(row)

    st.markdown(f"""
    <div style="border:1px solid #ddd; border-radius:10px; padding: 15px; margin:10px 0; background-color:#f9f9f9;">
        <h4>📅 {row['date'].strftime('%Y-%m-%d %H:%M')}</h4>
        <ul>
            <li><b>Appliance Usage:</b> {row['Appliances']:.2f} kWh</li>
            <li><b>Carbon Level:</b> {row['carbon_level']}</li>
            <li><b>Activity Detected:</b> {"Yes" if row['is_active'] else "No"}</li>
        </ul>
        <p>📋 <b>Recommendation:</b> <span style="color:green;">{rec}</span></p>
    </div>
    """, unsafe_allow_html=True)

```
